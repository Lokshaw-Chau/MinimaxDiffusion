```python
x = [1, 5, 10, 50, 100, 200, 500]
idc_y1 = [27.8, 49, 59.2, 85.3, 91.2, 93.6, 95.4]
idc_y2 = [22.8, 49, 63.3, 82.8, 88.6, 91.6, 93]


woof_y1 = [20.6, 33.4, 47.5, 75.4, 82.4, 87.4, 90]
woof_y2 = [20.8, 39.4, 49, 75.2, 82.2, 87, 89.6]
```

```python
import matplotlib.pyplot as plt

# Plot for ImageNet-IDC
plt.figure(figsize=(10, 5))
plt.plot(x, idc_y1, label='random')
plt.plot(x, idc_y2, label='minimax')
plt.xlabel('IPC')
plt.ylabel('acc')
plt.title('ImageNet-IDC')
plt.legend()
plt.show()

# Plot for ImageNet-WOOF
plt.figure(figsize=(10, 5))
plt.plot(x, woof_y1, label='random')
plt.plot(x, woof_y2, label='minimax')
plt.xlabel('IPC')
plt.ylabel('acc')
plt.title('ImageNet-WOOF')
plt.legend()
plt.show()
```
